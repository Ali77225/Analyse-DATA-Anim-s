3- Nettoyage des données

```python
# Copie pour éviter les warnings et garder un original
df_clean = df.copy()

# 0. Convertir Date en datetime (meilleure pratique)
df_clean['Date_Pub'] = pd.to_datetime(df_clean['Date_Pub'])

# 1. Supprimer les lignes avec valeurs manquantes (ici Produit)
df_clean = df_clean.dropna()

# 2. Supprimer les doublons
df_clean = df_clean.drop_duplicates()

# 3. Supprimer la colonne 'Studio'
df_clean = df_clean.drop(columns=['Studio'])

# 4. Supprimer la colonne 'Status7'
df_clean = df_clean.drop(columns=['Status'])

print("Données nettoyées ! ✅")
display(df_clean)
```

4- Enrichissement des données

```python
import pandas as pd
import numpy as np

# 1. Chargement des données (ENLEVEZ LE '#' au début de la ligne suivante)
df_clean = pd.read_csv('animes.csv')

# 2. Enrichissement des données
df_clean['Ecart'] = df_clean['Note_Meilleur_Ep'] - df_clean['Note_Pire_Ep']
df_clean['Regularite'] = 10 - df_clean['Ecart']
df_clean['Score_Qualite'] = (df_clean['Note_Globale'] * 0.6) + (df_clean['Regularite'] * 0.4)

# 3. Calcul de la Longueur (logarithme du nombre d'épisodes)
df_clean['Longueur'] = np.log(df_clean['Nb_Episodes'])

# 4. Affichage
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
# Affichage du tableau complet
display(df_clean)
```

5- ANALYSE STATISTIQUE

```python
ca_min = df_clean['Score_Qualite'].min()
ca_max = df_clean['Score_Qualite'].max()
ca_moyenne = df_clean['Score_Qualite'].mean()
ca_mediane = df_clean['Score_Qualite'].median()

print(f"le score_qualite minimal est: {ca_min:.2f}  | le score_qualite maximal est: {ca_max:.2f} ")
print(f"le score_qualite moyen est: {ca_moyenne:.2f}  | le mediane des score-qualite est: {ca_mediane:.2f} ")
```

Correlation

```python
import plotly.express as px

# 1. Calcul de la corrélation
correlation = df_clean[['Score_Qualite', 'Note_Globale', 'Regularite', 'Nb_Episodes']].corr()

# 2. Création de la figure (This is the missing step!)
fig = px.imshow(correlation, text_auto=True, aspect="auto", color_continuous_scale='RdBu_r')

# 3. Amélioration du design
fig.update_layout(
    title_text="Matrice de Corrélation",
    title_x=0.5,
    xaxis_title="Variables",
    yaxis_title="Variables",
    width=700,
    height=600
)

fig.show()
```

6- Analyse par genre

```python
import numpy as np

# 1. Agrégation des données (on utilise les noms exacts de ton CSV)
resume_genres = df_clean.groupby('Anime').agg({
    'Score_Qualite': 'sum',
    'Nb_Episodes': 'sum',
    'Note_Globale': 'mean'
}).sort_values(by='Score_Qualite', ascending=False)

# 2. Définition des conditions de multiplicateurs
conditions = [
    (resume_genres['Nb_Episodes'] <= 150),
    (resume_genres['Nb_Episodes'] > 150) & (resume_genres['Nb_Episodes'] <= 300),
    (resume_genres['Nb_Episodes'] > 300)
]

# Coefficients : x1.0 (0-150), x0.85 (150-300), x0.75 (>300)
multiplicateurs = [1.0, 0.85, 0.75]

# 3. Calcul de la Recommandation finale en base 100
# On applique le multiplicateur et on multiplie par 10 pour le format %
resume_genres['Recommandation_Pourcent'] = (resume_genres['Score_Qualite'] * np.select(conditions, multiplicateurs)) * 10

# 4. Affichage : On ne sélectionne QUE les colonnes voulues pour l'affichage final
colonnes_finales = ['Score_Qualite', 'Nb_Episodes', 'Note_Globale', 'Recommandation_Pourcent']

display(resume_genres[colonnes_finales].style.format({
    'Score_Qualite': '{:.1f}',
    'Nb_Episodes': '{:g}',
    'Note_Globale': '{:.2f}',
    'Recommandation_Pourcent': '{:.1f} %'
}))
```

7- Visualisation

```python
import matplotlib.pyplot as plt

# 7- Visualisation
score_par_anime = df_clean.groupby('Anime')['Score_Qualite'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
# Utilisation de la couleur choisie
score_par_anime.plot(kind='bar', color='blue')

plt.title("Score Qualité par Anime")
plt.ylabel('SQ(/10)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
